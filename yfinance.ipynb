```python
import matplotlib.pyplot as plt
import seaborn as sns
import japanize_matplotlib
fig = plt.figure(figsize=(15,6))

for i, feature in enumerate(["Close","High","Low","Open","Volume"]):
    ax = plt.subplot(3,2,i+1)
    ax.set_title(feature)
    ohlcv[feature].tail(500).plot(ax=ax)
plt.tight_layout()
```

```python
ohlcv["ret"] = (ohlcv["Close"] - ohlcv["Open"]) / ohlcv["Open"]
```

```python
ohlcv
```

```python
plt.title("任天堂 日次リターン分布")
ohlcv["ret"].plot(kind="hist",bins=20)
plt.yscale("log")
```

```python
SP500 = yf.Ticker("^GSPC")

ohlcv_SP500 = yf.download("^GSPC",start="2020-01-01",end="2024-12-31", interval="1d")
ohlcv_SP500
```

```python
#multi indexでなくする
ohlcv_SP500_single = ohlcv_SP500.copy()
ohlcv_SP500_single.columns = ohlcv_SP500_single.columns.droplevel(1)
ohlcv_SP500_single
```

```python
ohlcv_SP500["retSP500"] = (ohlcv_SP500["Close"] - ohlcv_SP500["Open"]) / ohlcv_SP500["Open"]
```

```python
ohlcv_SP500
```

```python
plt.title("SP500 日次リターン")
ohlcv_SP500["retSP500"].plot(kind="hist",bins=20,alpha=0.5,label = "SP500")
ohlcv["ret"].plot(kind="hist",bins=20,alpha=0.5,label="nintendo")
plt.yscale("log")
plt.legend()
plt.plot()
```

```python
import pandas as pd
merged_df = ohlcv.merge(ohlcv_SP500,on="Date",how="left")
merged_df
```

```python
import numpy as np
strategy1 = merged_df.copy()
strategy1["retSP500_sign"] = np.sign(strategy1["retSP500"])
strategy1["retSP500_sign_s1"] = strategy1["retSP500_sign"].shift(1)
```

```python
plt.title("SP500リターン (shift=1)")
strategy1.groupby("retSP500_sign_s1").size().plot(kind="bar")
strategy1.columns
```

```python
strategy1["signal"] = 0
strategy1.loc[strategy1["retSP500_sign_s1"] == 1, "signal"] = 1
strategy1.loc[strategy1["retSP500_sign_s1"] == -1, "signal"] = -1
```

```python
strategy1
```

```python
fig = plt.figure(figsize=(10,3))
plt.title("各戦略の累積リターン")
plt.plot((strategy1["signal"]*strategy1["ret"]).cumsum(),label="戦略1")
plt.legend()
```

```python
fig = plt.figure(figsize=(10,3))
plt.title("各戦略のリターン")
plt.plot(strategy1["ret"].cumsum(),label="すべてロング",c="lightgray")
plt.plot((strategy1["signal"]*strategy1["ret"]).cumsum(),label="戦略1")
plt.plot((strategy1["signal"].replace(-1,0)*strategy1["ret"]).cumsum(),label="戦略１ ロングのみ")
plt.plot((strategy1["signal"].replace(1,0)*strategy1["ret"]).cumsum(),label="戦略１ ショートのみ")
plt.legend()
```
